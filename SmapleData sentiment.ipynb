```python
from textblob import TextBlob
import math
import re
```

```python
def clean_tweet(tweet):
    '''
    Utility function to clean the text in a tweet by removing 
    links and special characters using regex.
    '''
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
```

```python
def analize_sentiment(tweet):
    '''
    Utility function to classify the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
SA = np.zeros(106)
Sentiment = np.zeros(106)
# create a new column that represent the result of sentiment analysis
for i in range(106):
    if df['text'][i] != df['text'][i]:
        continue
    else:
        SA[i] = analize_sentiment(df['text'][i])
        Sentiment[i] = TextBlob(clean_tweet(df['text'][i])).sentiment.polarity
        #print(i)
```
